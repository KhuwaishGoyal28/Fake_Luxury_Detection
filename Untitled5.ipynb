```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Set up the figure
fig, ax = plt.subplots(figsize=(10, 12))
ax.set_xlim(0, 10)
ax.set_ylim(0, 12)
ax.axis('off')  # Hide axes

# Box positions and labels
components = [
    ("1. User Interface Layer\n(Streamlit UI)", 11),
    ("2. Input Structuring Layer\n(Python Dicts / Lists)", 9.5),
    ("3. Visualization Layer\n(Matplotlib + NumPy)", 8),
    ("4. Media Embedding Layer\n(QR Code via PIL + MPL)", 6.5),
    ("5. Output Generation Layer\n(PNG with MPL, PDF with PIL)", 5),
    ("6. Output Preview & Download\n(Streamlit)", 3.5)
]

# Draw each box and connector arrow
for i, (text, y) in enumerate(components):
    # Draw box
    ax.add_patch(Rectangle((2, y - 0.7), 6, 1.2, edgecolor='black', facecolor='#e6f2ff'))
    # Add text
    ax.text(5, y, text, ha='center', va='center', fontsize=10)
    # Add arrow (except last)
    if i < len(components) - 1:
        next_y = components[i + 1][1]
        ax.annotate('', xy=(5, next_y + 0.5), xytext=(5, y - 0.6),
                    arrowprops=dict(facecolor='black', arrowstyle='->'))

# Save the diagram
plt.title("Architecture Diagram - Resume Builder Project", fontsize=14, weight='bold')
plt.savefig("resume_builder_architecture.png", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_detailed_architecture_diagram():
    fig, ax = plt.subplots(figsize=(12, 9))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 13)
    ax.axis('off')

    # Define layer boxes with position and size
    layers = [
        ("User Interface Layer", "Streamlit-based form inputs, section config", 10.5, "#5DADE2"),
        ("Data Handling Layer", "Python dicts/lists structure resume sections", 8.0, "#85C1E9"),
        ("Visualization Engine", "Matplotlib renders text, sections, QR", 5.5, "#AED6F1"),
        ("File Generation & Output Layer", "Export to PNG and convert to PDF", 3.0, "#D6EAF8")
    ]

    # Draw rectangles and annotations
    for name, desc, y, color in layers:
        rect = Rectangle((2, y, 6, 2), facecolor=color, edgecolor='black', linewidth=1.5)
        ax.add_patch(rect)
        ax.text(5, y + 1.4, name, ha='center', va='center', fontsize=13, weight='bold', color='#154360')
        ax.text(5, y + 0.6, desc, ha='center', va='center', fontsize=10, color='#1B4F72')

    # Multiple arrows between sub-modules
    arrows = [
        ((5, 10.5), (5, 9.9)),  # UI -> Data Handling
        ((4.5, 10.5), (4.5, 9.9)),
        ((5.5, 10.5), (5.5, 9.9)),

        ((5, 8.0), (5, 7.4)),  # Data -> Visualization
        ((4.5, 8.0), (4.5, 7.4)),
        ((5.5, 8.0), (5.5, 7.4)),

        ((5, 5.5), (5, 4.9)),  # Visualization -> Output
        ((4.5, 5.5), (4.5, 4.9)),
        ((5.5, 5.5), (5.5, 4.9))
    ]

    for start, end in arrows:
        ax.annotate('', xy=end, xytext=start, arrowprops=dict(facecolor='gray', arrowstyle='->', lw=1.2))

    # Title
    ax.text(5, 12.5, "Resume Builder - Multi-Path Architecture Diagram", ha='center',
            fontsize=16, weight='bold', color='#154360')

    # Save and display
    plt.tight_layout()
    plt.savefig("/mnt/data/multi_connection_architecture_diagram.png", dpi=300)
    plt.show()

draw_detailed_architecture_diagram()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_enhanced_architecture_diagram():
    fig, ax = plt.subplots(figsize=(14, 12))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 16)
    ax.axis('off')

    # Define extended layers and new components
    layers = [
        ("User Interface Layer", "Streamlit Inputs, Upload, Preview, Download", 14, "#5DADE2"),
        ("Input Structuring Layer", "Data cleaned & stored in dict/list", 12, "#85C1E9"),
        ("Validation Layer", "Check for required fields, formats, injection", 10, "#A9CCE3"),
        ("Data Transformation Layer", "Split text, sanitize, parse skills etc.", 8, "#D6EAF8"),
        ("Visualization Engine", "Matplotlib layout, sections, skill bars", 6, "#AED6F1"),
        ("Media Embedding Layer", "QR Code, logos, photos via PIL", 4.2, "#D1F2EB"),
        ("File Generation Layer", "Export to PNG & convert to PDF", 2.6, "#D5F5E3"),
        ("Output Layer", "Preview + Download (PDF)", 1.0, "#FADBD8")
    ]

    # Draw boxes and text
    for name, desc, y, color in layers:
        rect = Rectangle((2.5, y), 7, 1.5, facecolor=color, edgecolor='black', linewidth=1.5)
        ax.add_patch(rect)
        ax.text(6, y + 1.0, name, ha='center', va='center', fontsize=12, weight='bold', color='#154360')
        ax.text(6, y + 0.4, desc, ha='center', va='center', fontsize=10, color='#1B4F72')

    # Draw arrows
    for i in range(len(layers) - 1):
        y_start = layers[i][2]
        y_end = layers[i + 1][2] + 1.5
        ax.annotate('', xy=(6, y_end), xytext=(6, y_start),
                    arrowprops=dict(facecolor='gray', arrowstyle='->', lw=1.2))

    # Title
    ax.text(6, 15.5, "Enhanced Architecture Diagram – Resume Builder", ha='center',
            fontsize=16, weight='bold', color='#154360')

    # Save and show
    plt.tight_layout()
    plt.savefig("enhanced_resume_architecture_diagram.png", dpi=300)
    plt.show()

draw_enhanced_architecture_diagram()
```
